we are going to create a new variable 'purpose_type' related with the column 'purpose'. The difference is that this new column will contain a certain word from 'purpose' column which has many words. 

```python
import nltk
from nltk.stem import WordNetLemmatizer

def lemmatize_purpose(purpose):
    word_lemma= WordNetLemmatizer()
    text=purpose
    words= nltk.word_tokenize(text)
    lemma=[word_lemma.lemmatize(w, pos = 'n') for w in words]
    return lemma
```

```python
def grouping_purpose(lemmas):
    
    wedding= ['wedding']
    assets= ['transaction','real state','property']
    house= ['house']
    car= ['car']
    education= ['education']
    text= lemmatize_purpose(lemmas)
    
    for i in wedding:
        if i in text:
            return 'wedding'
    for j in assets:
        if j in text:
            return 'assets'
    for k in house:
        if k in text:
            return 'house'
    for l in car:
        if l in text:
            return 'car'
    for m in education:
        if m in text:
            return 'education'
    
```

```python
df['purpose_type']= df['purpose'].apply(grouping_purpose)
df['purpose_type'].value_counts()
```

#### Once we prove the existance of missing values  it is time to replace 

quantitative variables

```python
df['balconies']= df['balconies'].fillna(0)
```

```python
df[['kitchen_area','living_area','ceiling_height', 'floors_total']].describe()
```

replacing values acording to kitchen_area + living_area <= total_area

```python
avg_ratio= (df['living_area']/df['total_area']).mean()
if df['kitchen_area'].median()>df['total_area'].any():
    df['kitchen_area']= df['kitchen_area'].fillna(avg_ratio)
else:
     df['kitchen_area']= df['kitchen_area'].fillna(df['kitchen_area'].median())
```

```python
rooms_to_living_area = df.groupby("bedrooms")["living_area"].mean().to_dict()
 
def filler(row):
    if np.isnan(row["living_area"]):
        return rooms_to_living_area[row["bedrooms"]]
    return row["living_area"]
 
df["living_area"] = df.apply(filler, axis=1)
```

```python
if df['floor'].median()>df['floors_total'].any():
    df['floor']= df['floor'].fillna(df['floors_total'].median())
else:
    df['floor']= df['floor'].fillna(df['floor'].median())
```

```python
df['ceiling_height']= df['ceiling_height'].fillna(df['ceiling_height'].mean())
```
